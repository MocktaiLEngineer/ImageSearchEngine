```python
import numpy as np
import torch
import pandas as pd
import glob
import Config
```

```python
emb = np.load('../models/H&M-Embeddings.npy')
```

```python
articles = pd.read_csv(r'../data/raw/h-and-m-personalized-fashion-recommendations/articles.csv')
img_path = glob.glob('../data/raw/h-and-m-personalized-fashion-recommendations/images/*/*')
```

```python
embeddings_df = pd.DataFrame(emb, columns=[f"{i}" for i in range(config.embedding_size)])
```

```python
embeddings_df
```

```python
embeddings_df['embeddings'] = embeddings_df.apply(lambda row: torch.tensor(row.values), axis=1)

# Drop the original float columns
embeddings_df = embeddings_df.drop([f"{i}" for i in range(config.embedding_size)], axis=1)
```

```python
embeddings_df['image_path'] = img_path
```

```python
def extractImageId(image_path):
    image_filename = image_path.split('/')[-1]
    image_id = image_filename.rstrip('.jpg')
    image_id = image_id[1:]
    return image_id
    
embeddings_df['image_id'] = embeddings_df['image_path'].apply(lambda x: extractImageId(x))
```

```python
# Drop the image path column
embeddings_df = embeddings_df.drop(['image_path'], axis=1)
```

```python
embeddings_df.to_csv(r'../models/')
```
